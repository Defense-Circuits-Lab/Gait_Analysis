### 1. Create template of synchro motif

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
```

```python
class MotifTemplate:
    
    def __init__(self, led_on_time_in_ms: int, on_off_period_length_in_ms: int, motif_duration_in_ms: int):
        self.led_on_time_in_ms = led_on_time_in_ms
        self.on_off_period_length_in_ms = on_off_period_length_in_ms
        self.motif_duration_in_ms = motif_duration_in_ms
        self.template_timeseries = self._generate_template_timeseries()
        
    
    def _generate_template_timeseries(self) -> np.ndarray:
        led_on_off_period = np.zeros((self.on_off_period_length_in_ms), dtype='uint8')
        led_on_off_period[1:self.led_on_time_in_ms+1] = 1
        full_repetitions = self.motif_duration_in_ms // self.on_off_period_length_in_ms
        remaining_ms = self.motif_duration_in_ms % self.on_off_period_length_in_ms
        motif_template = np.concatenate([led_on_off_period]*(full_repetitions+1))
        adjusted_end_index = self.on_off_period_length_in_ms*full_repetitions + remaining_ms
        return motif_template[:adjusted_end_index]
    
    
    def adjust_template_timeseries_to_fps(self, fps: int) -> None:
        fps_adjusted_templates = []
        framerate = fps/1000
        max_frames = int(self.template_timeseries.shape[0] * framerate)
        max_offset = 1000 // fps
        for offset_in_ms in range(max_offset):
            image_timestamps = np.linspace(0+offset_in_ms, self.template_timeseries.shape[0]+offset_in_ms, max_frames, dtype='int')
            while image_timestamps[-1] >= self.template_timeseries.shape[0]:
                image_timestamps = image_timestamps[:-1]
            adjusted_template = self.template_timeseries[image_timestamps].copy()
            fps_adjusted_templates.append((adjusted_template, offset_in_ms))
        self.fps_adjusted_templates = fps_adjusted_templates
```

```python
class SessionTemplate:
    
    def __init__(self) -> None:
        self.motif_templates = []
    
    def add_motif_template(self, motif_template: MotifTemplate) -> None:
        self.motif_templates.append(motif_template)
        self._update_whole_session_template()
        
    
    def _update_whole_session_template(self) -> None:
        individual_motif_template_timeseries = [elem.template_timeseries for elem in self.motif_templates]
        self.whole_session_template = np.concatenate(individual_motif_template_timeseries)
        
        
    def adjust_template_timeseries_to_fps(self, fps: int) -> List[np.ndarray]:
        fps_adjusted_templates = []
        framerate = fps/1000
        max_frames = int(self.whole_session_template.shape[0] * framerate)
        max_offset = 1000 // fps
        for offset_in_ms in range(max_offset):
            image_timestamps = np.linspace(0+offset_in_ms, self.whole_session_template.shape[0]+offset_in_ms, max_frames, dtype='int')
            while image_timestamps[-1] >= self.whole_session_template.shape[0]:
                image_timestamps = image_timestamps[:-1]
            adjusted_template = self.whole_session_template[image_timestamps].copy()
            fps_adjusted_templates.append((adjusted_template, offset_in_ms))
        return fps_adjusted_templates
```
