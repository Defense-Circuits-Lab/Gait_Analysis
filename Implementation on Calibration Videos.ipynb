```python
import numpy as np
from aniposelib.boards import CharucoBoard, Checkerboard
from aniposelib.cameras import Camera, CameraGroup
from aniposelib.utils import load_pose2d_fnames
from cv2 import aruco
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vidnames = [['220719_test01_bottom_defisheye.mp4'],
            ['220719_test01006.AVI'],
            ['220719_test01_side1.mp4'],
            ['220719_test01_side2.mp4']]

cam_names = ['Bottom', 'Top', 'Side1', 'Side2']

n_cams = len(vidnames)

board = CharucoBoard(5, 7,
                     square_length=1, # here, in mm but any unit works
                     marker_length=0.8,
                     marker_bits=6, aruco_dict=aruco.DICT_6X6_250)
#5, 7, square_length = 1, marker_length = 0.8, marker_bits = 6, aruco_dict=aruco.DICT_6X6_250
#https://mecaruco2.readthedocs.io/en/latest/notebooks_rst/Aruco/sandbox/ludovic/aruco_calibration_rotation.html
#https://docs.rs/opencv/latest/opencv/aruco/enum.PREDEFINED_DICTIONARY_NAME.html
        
# the videos provided are fisheye, so we need the fisheye option
cgroup = CameraGroup.from_names(cam_names)
```

```python
size = (2000, 2000)
board_image = board.draw(size)
fig = plt.Figure()
image = plt.imshow(board_image)
```

```python
cgroup.calibrate_videos(vidnames, board)
```
